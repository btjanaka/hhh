# Feature Visualization

Visualize the features for one audio file.

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import librosa
import librosa.display

sns.set()  # Aesthetic plots
```

```python
# Extract features.
audio, sample_rate = librosa.load("../data/bird-audio-detection/ff1010-wav/127833.wav", res_type="kaiser_fast")
mfcc = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
chromagram = librosa.feature.chroma_stft(audio)
spectal_contrast = librosa.feature.spectral_contrast(audio, sr=sample_rate)
```

```python
plt.subplots(3, figsize=(10, 12))

plt.subplot(3,1,1)
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title('MFCC')

plt.subplot(3,1,2)
librosa.display.specshow(chromagram, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chromagram')

plt.subplot(3,1,3)
librosa.display.specshow(spectal_contrast, x_axis='time')
plt.colorbar()
plt.ylabel('Frequency bands')
plt.title('Spectral contrast')

plt.tight_layout()
plt.savefig("features.png")
plt.show()
```
